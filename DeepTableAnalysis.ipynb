```python
# media dos campeoes

champions_points = []

champions_points.append(firsts_serie_a_22.iloc[0]['PTS'])
champions_points.append(firsts_serie_a_23.iloc[0]['PTS'])
champions_points.append(firsts_serie_a_24.iloc[0]['PTS'])
champions_points.append(firsts_serie_a_25.iloc[0]['PTS'])
champions_points.append(firsts_serie_a_26.iloc[0]['PTS'])
champions_points.append(firsts_serie_a_27.iloc[0]['PTS'])
champions_points.append(firsts_serie_a_28.iloc[0]['PTS'])
champions_points.append(firsts_serie_a_29.iloc[0]['PTS'])
champions_points.append(firsts_serie_a_30.iloc[0]['PTS'])
```
