```python
# Select all centre forwards

attackers = df.loc[df['player_position'] == 'CF']
```

```python
att = attackers.groupby('player_name').agg(list)
```

```python
# Separate stats by position

gk_stats = ["matches", "clean_sheets","defenses","difficult_defenses","goals_conceded"]
def_stats = ["matches", "clearances", "stolen_balls", "clean_sheets", "passes", "wrong_passes", "intercepted_passes", "tackles", "fouls_committed"]
att_stats = ["matches", "goals", "assists", "shots", "shots_on_target" ]

grouper = ["player_name", "player_position", "season", "club_owner"]
```

```python
df.groupby(["player_name", "player_position"])
```

```python
# Get keeper stats

gks = df.loc[df['player_position'] == 'GK']
```

```python
gks = gks.groupby(grouper)[gk_stats].sum()
```

```python
gks.sort_values(by='matches', ascending=False)
```

```python
# Center Backs

cbs = df.loc[df['player_position'] == 'CB' ]
```

```python
cbs = cbs.groupby(grouper)[def_stats].sum()
```

```python
cbs.sort_values(by=['matches','tackles','intercepted_passes','stolen_balls','passes'], ascending=False)
```

```python
# Right Backs 

rbs = df.loc[df['player_position'] == 'RB' ]
```

```python
rbs = rbs.groupby(grouper)[def_stats].sum()
```

```python
rbs.sort_values(by=['matches','passes','intercepted_passes'], ascending=False)
```

```python
# Center Forward

cfs = df.loc[df['player_position'] == 'CF' ]
```

```python
cfs = cfs.groupby(grouper)[att_stats].sum()
```

```python
cfs.sort_values(by=['matches','goals','assists'], ascending=False)
```
